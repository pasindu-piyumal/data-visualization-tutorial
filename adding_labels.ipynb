```python
import matplotlib.pyplot as plt
```

```python
def add_labels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha='center')

x = ['FAGS','FAPS','FOC','FGS','FSS','FMS']
y = [330,350,330,400,550,560]

plt.bar(x,y)
add_labels(x,y)

plt.title('Faculty Admission')
plt.xlabel('Faculties')
plt.ylabel('Number of admissions')
```

```python
def add_labels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha='center', bbox=dict(facecolor='red', alpha=0.8))

x = ['FAGS','FAPS','FOC','FGS','FSS','FMS']
y = [330,350,330,400,550,560]

plt.bar(x,y)
add_labels(x,y)

plt.title('Faculty Admission')
plt.xlabel('Faculties')
plt.ylabel('Number of admissions')
```

## Add grid lines

```python
import numpy as np
```

```python
x1 = np.linspace(0,5,100)
y1 = np.cos(2*np.pi*x1) * np.exp(-x1)

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(12,5))

ax2.plot(x1, y1)
ax2.set_title("plot with grid")
ax2.grid(True)
plt.show()
```

## Creating subplots

```python
fig, ax = plt.subplots(2,2)

x= np.linspace(0,10,1000)

ax[0,0].plot(x, np.sin(x), 'r-', label='sin(x)')
ax[0,1].plot(x, np.cos(x), 'g--', label='cos(x)')
ax[1,0].plot(x, np.tan(x), 'y+', label='tan(x)')
ax[1,1].plot(x, np.sinc(x), 'c.-', label='sinc(x)')

for axes in ax.flat:
    axes.legend()
plt.tight_layout()
plt.show()
```

```python
x= np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig, axes = plt.subplots(nrows=1,ncols=3, figsize=(12,4))

axes[0].plot(x, y1, color='blue', label='sin(x)')
axes[1].plot(x, y2, color='red', label='cos(x)')
axes[2].plot(x, y3, color='green', label='tan(x)')

axes[0].set_title('sin function')
axes[1].set_title('cos function')
axes[2].set_title('tan function')

for ax in axes:
    plt.legend()
plt.tight_layout()
plt.show()
```
